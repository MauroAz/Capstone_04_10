```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
## categorical variable NEEDs to be wraped in :::c()::::
linreg = smf.ols(formula='TARGET ~ C(CODE_GENDER) + AMT_INCOME_TOTAL + AMT_CREDIT + AMT_ANNUITY + AMT_GOODS_PRICE + REGION_POPULATION_RELATIVE + EXT_SOURCE_1 + EXT_SOURCE_2 + EXT_SOURCE_3 + APARTMENTS_AVG + BASEMENTAREA_AVG + YEARS_BEGINEXPLUATATION_AVG + YEARS_BUILD_AVG + COMMONAREA_AVG + ENTRANCES_AVG + ELEVATORS_AVG + LANDAREA_AVG + LIVINGAPARTMENTS_AVG + LIVINGAREA_AVG + NONLIVINGAPARTMENTS_AVG + NONLIVINGAREA_AVG + APARTMENTS_MODE + YEARS_BEGINEXPLUATATION_MODE + YEARS_BUILD_MODE + COMMONAREA_MODE + ELEVATORS_MODE + ENTRANCES_MODE + FLOORSMAX_MODE + FLOORSMIN_MODE + LANDAREA_MODE + LIVINGAPARTMENTS_MODE + LIVINGAREA_MODE + NONLIVINGAPARTMENTS_MODE + APARTMENTS_MEDI + BASEMENTAREA_MEDI + YEARS_BEGINEXPLUATATION_MEDI + YEARS_BUILD_MEDI + COMMONAREA_MEDI + ELEVATORS_MEDI + FLOORSMAX_MEDI + FLOORSMIN_MEDI + LANDAREA_MEDI + LIVINGAPARTMENTS_MEDI + LIVINGAREA_MEDI + NONLIVINGAPARTMENTS_MEDI + NONLIVINGAREA_MEDI', data=LinRegression).fit()

```
